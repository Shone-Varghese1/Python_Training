```python
df.select("city").distinct().show()
```

#Display only customer_id, name, and annual_spend columns.

```python
df.select("customer_id","name","annual_spend").show()
```

#Filter all customers who spend more than 60000 annually.

```python
df.filter(df.annual_spend>60000).show()
```

#Show all customers from Delhi who are younger than 30.

```python
df.filter((df.city=="Delhi") & (df.age<30)).show()
```

#Create a new column named "spend_lakh" = annual_spend / 100000.

```python
from pyspark.sql.functions import col,when
df2=df.withColumn("spend_lakh",col("annual_spend")/100000)
df2.show()
```

Create a new column "customer_type"
Logic:
spend > 70000 → Premium
else → Standard
Use when() and otherwise().

```python

df = df.withColumn("customer_type", when(col("annual_spend") > 70000, "Premium").otherwise("Standard"))

df.show()
```

#Show customers whose name starts with the letter A.

```python

df.filter(col("name").startswith("A")).show()
```

#Filter customers where category is either Clothing or Electronics.

```python
df.filter((col("category")=="Clothing") |(col("category")== "Electronics")).show()
```

#Convert the city name to uppercase using the upper() function.

```python
from pyspark.sql.functions import upper
df3=df.withColumn("city",upper(col("city")))
df3.show()
```

#Remove the category column from the DataFrame.

```python
df3.drop("category").show()
```

#Sort customers by age in descending order.

```python

from pyspark.sql.functions import desc

df.orderBy(desc("age")).show()
```

#Find customers who are younger than the average age of the entire dataset.

```python


from pyspark.sql.functions import avg, col

avg_age = df.select(avg("age")).first()[0]
print(avg_age)

df.filter(col("age") < avg_age).show()

```

#Display the top 5 highest spending customers.
