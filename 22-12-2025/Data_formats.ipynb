```python
from pyspark.sql import SparkSession



spark=SparkSession.builder.getOrCreate()
```

```python
data = [
    ("ORD001","Delhi","Laptop",45000,"2024-01-05"),
    ("ORD002","Mumbai","Mobile",32000,"2024-01-06"),
    ("ORD003","Bangalore","Tablet",30000,"2024-01-07"),
    ("ORD004","Delhi","Laptop",55000,"2024-01-08"),
    ("ORD005","Mumbai","Tablet",34000,"2024-01-09")
]
```

```python
columns=["order_id","city","product","price","order_date"]
df=spark.createDataFrame(data,columns)
df.show()
```

```python
df.write.mode("overwrite").parquet("orders.parquet")
```

```python
df_parquet=spark.read.parquet("orders.parquet")
df_parquet.show()
```

```python
df.write.mode("overwrite").orc("orders.orc")
```

```python
df_orc=spark.read.orc("orders.orc")
df_orc.show()
```

#parquet-column wise compression good for aggregation ques

#parquet fails when comes to large historical data like telecom,insurance
orc is highly compressed, less storage cost
